```python
import random
import pandas as pd
```

```python
data = []
def generate (catStart,catStop,classStart,numOfClassesPerCat):

  i = catStart
  classId = classStart
  while i != catStop + 1:
    for x in range(0,numOfClassesPerCat):
      localClass = classId
      period = random.randint(0,6)
      remaining_seats = random.randint(0,6)
      data.append([i,localClass, period,remaining_seats,"","",""])
      classId += 1
    i += 1
```

```python
generate(0,5,0,5)
```

```python
len(data)
```

```python
df = pd.DataFrame(data,columns=["courseID","classID","period","available_seats","name","priority","prerequisites"])
```
